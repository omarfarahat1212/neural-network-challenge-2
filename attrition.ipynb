## Part 1: Preprocessing

```python
# Import our dependencies
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import pandas as pd
import numpy as np
from tensorflow.keras.models import Model
from tensorflow.keras import layers

#  Import and read the attrition data
attrition_df = pd.read_csv('https://static.bc-edx.com/ai/ail-v-1-0/m19/lms/datasets/attrition.csv')
attrition_df.head()
```

```python
# Determine the number of unique values in each column.
attrition_df.nunique()
```

```python
attrition_df.columns
```

```python
# Create y_df with the Attrition and Department columns
y_df = attrition_df[["Attrition","Department"]]
y_df
```

```python
# Create a list of at least 10 column names to use as X data
x_list = ['Age',  'BusinessTravel',  'DistanceFromHome',
       'Education', 'EducationField', 'EnvironmentSatisfaction', 'HourlyRate',
       'JobInvolvement', 'JobLevel', 'JobRole', 'JobSatisfaction',
       'MaritalStatus', 'NumCompaniesWorked', 'OverTime', 'PercentSalaryHike',
       'PerformanceRating', 'RelationshipSatisfaction', 'StockOptionLevel',
       'TotalWorkingYears', 'TrainingTimesLastYear', 'WorkLifeBalance',
       'YearsAtCompany', 'YearsInCurrentRole', 'YearsSinceLastPromotion',
       'YearsWithCurrManager']


# Create X_df using your selected columns
X_df = attrition_df[x_list]

# Show the data types for X_df
X_df
```

```python
# Split the data into training and testing sets
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X_df, y_df, test_size=0.2, random_state=42)
```

```python
X_df.info()
```

```python
# Filter columns with object dtype
object_columns = X_df.select_dtypes(include=['object'])

# Display the count of unique values for object type columns
print(object_columns.nunique())
```

```python
# Convert your X data to numeric data types however you see fit
# Add new code cells as necessary

# Select object type columns
object_columns_train = X_train.select_dtypes(include=['object'])
object_columns_test= X_test.select_dtypes(include=['object'])
# One-hot encode object type columns
X_train_encoded = pd.get_dummies(object_columns_train, drop_first=True)
X_test_encoded = pd.get_dummies(object_columns_test, drop_first=True)

```

```python
# Create a StandardScaler
scaler = StandardScaler()


# Fit the StandardScaler to the training data
scaler.fit(X_train_encoded)

# Scale the training and testing data
X_train_scaled = scaler.transform(X_train_encoded)
X_test_scaled = scaler.transform(X_test_encoded)

X_train_scaled = np.array(X_train_scaled)
X_train_scaled
```

```python
# Create one-hot encoding for the 'Department' column using pd.get_dummies()
encoded_department_train = pd.get_dummies(y_train['Department'])
encoded_department_test = pd.get_dummies(y_test['Department'])

# Convert the encoded DataFrames to NumPy arrays if needed
y_train_encoded_department = encoded_department_train.to_numpy()
y_test_encoded_department = encoded_department_test.to_numpy()

# Print the encoded 'Department' column for training data
print("Encoded Department column for training data:")
print(y_train_encoded_department)

# Print the encoded 'Department' column for testing data
print("\nEncoded Department column for testing data:")
print(y_test_encoded_department)
```

```python
# Create one-hot encoding for the 'Attrition' column using pd.get_dummies()
encoded_attrition_train = pd.get_dummies(y_train['Attrition'])
encoded_attrition_test = pd.get_dummies(y_test['Attrition'])

# Convert the encoded DataFrames to NumPy arrays if needed
y_train_encoded_attrition = encoded_attrition_train.to_numpy()
y_test_encoded_attrition = encoded_attrition_test.to_numpy()

# Print the encoded 'Attrition' column for training data
print("Encoded Attrition column for training data:")
print(y_train_encoded_attrition)

# Print the encoded 'Attrition' column for testing data
print("\nEncoded Attrition column for testing data:")
print(y_test_encoded_attrition)

```

## Create, Compile, and Train the Model

```python
import tensorflow as tf

# Find the number of columns in the X training data

num_input_features = X_train_encoded.shape[1]
print("Number of columns in the X training data:", num_input_features)

# Create the input layer
input_layer = tf.keras.layers.Input(shape=(num_input_features,))

# Create at least two shared layers
shared_layer_1 = tf.keras.layers.Dense(64, activation='relu')(input_layer)
shared_layer_2 = tf.keras.layers.Dense(32, activation='relu')(shared_layer_1)
shared_layer_3 = tf.keras.layers.Dense(16, activation='relu')(shared_layer_2)
```

```python
# Create a branch for Department
# with a hidden layer and an output layer

# Create the hidden layer
hidden_layer_dept = tf.keras.layers.Dense(64, activation='relu')(shared_layer_3)

# Create the output layer
output_layer_dept = tf.keras.layers.Dense(3, activation='softmax')(hidden_layer_dept)

```

```python
# Create a branch for Attrition
# with a hidden layer and an output layer

# Create the hidden layer
hidden_layer_attrition = tf.keras.layers.Dense(64, activation='relu')(shared_layer_3) 

# Create the output layer
output_layer_attrition = tf.keras.layers.Dense(2, activation='softmax')(hidden_layer_attrition)
```

```python
# Create the model
model = tf.keras.models.Model(inputs=input_layer, outputs=[output_layer_dept, output_layer_attrition])


# Compile the model with metrics for each output
model.compile(optimizer='adam',
              loss=['categorical_crossentropy', 'categorical_crossentropy'],
              metrics=['accuracy', 'accuracy'])



# Summarize the model
model.summary()
```

```python
# Print shapes for debugging
print("X_train_scaled shape:", X_train_scaled.shape)
print("y_train_encoded_department shape:", y_train_encoded_department.shape)
print("y_train_encoded_attrition shape:", y_train_encoded_attrition.shape)

print("X_train_scaled:", X_train_scaled)
print("y_train_encoded_department:", y_train_encoded_department)
print("y_train_encoded_attrition:", y_train_encoded_attrition)




```

```python
# Train the model
model.fit(X_train_scaled, [y_train_encoded_department, y_train_encoded_attrition], epochs=100)
```

```python
# Evaluate the model for department predictions
loss_department, accuracy_department = model.evaluate(X_test_scaled, y_test_encoded_department, verbose=0)

# Evaluate the model for attrition predictions
loss_attrition, accuracy_attrition = model.evaluate(X_test_scaled, y_test_encoded_attrition, verbose=0)
```

```python
# Print the accuracy for both department and attrition

# Print the accuracy for department prediction
print("Accuracy for Department prediction:", accuracy_department)

# Print the accuracy for attrition prediction
print("Accuracy for Attrition prediction:", accuracy_attrition)
```
